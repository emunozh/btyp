```python
def getData(file_name, aggregate="Merge", mode = "W"):
    file_path = os.path.join(os.getcwd(), "Data", "ResultDC_{}_{}.csv".format(mode, file_name))
    data_raw = pd.read_csv(file_path, index_col=0)
    data_raw.columns = [col.strip() for col in data_raw.columns]
    if aggregate == "Merge":
        pieces = []
        for i in range(999):
            dat = data_raw[["Residential", "Consumption", "Demand_{}".format(i)]]
            dat.columns = ["Residential", "Consumption", "Demand"]
            pieces.append(dat.astype(float))
        data_out = pd.concat(pieces)
    elif aggregate == "Mean":
        demand = data_raw.loc[:, "Demand_0":"Demand_998"].astype(float)
        demand = demand.dropna().mean(1)
        data_out = data_raw[["Residential", "Consumption"]]
        data_out["Demand"] = demand
    else:
        data_out = data_raw
    return(data_out)
```

```python
def plotTyp(typ_data, typ_name, lim=1e7):
    fig, ax = plt.subplots(figsize=(6,6))
    lim_s = [typ_data['Demand'].iloc[a] < lim and typ_data['Consumption'].iloc[a] < lim for a in range(typ_data.shape[0])]
    ax.scatter(
        typ_data['Demand'][lim_s],
        typ_data['Consumption'][lim_s],
        s=typ_data['Residential'][lim_s]*100, alpha=0.7, c=(0.1,0.1,0.1))
    ax.xaxis.get_major_formatter().set_powerlimits((0, 1))
    ax.yaxis.get_major_formatter().set_powerlimits((0, 1))
    x1 = np.arange(min(typ_data['Demand'][lim_s]),
                   max(typ_data['Demand'][lim_s]), 10)
    ax.plot(x1, x1, linestyle='--', color="red", linewidth=1)
    str_lim = "{:.0e}".format(lim).split("+")[-1]
    fig.savefig("FIGURES/{}_{}.png".format(typ_name, str_lim))
```

# 5. Cuantify the performance

```python
def truncateData(typ_data, lim):
    typ_data = typ_data[
        (typ_data.Demand < lim) &\
        (typ_data.Consumption < lim) &\
        ~typ_data.Demand.isnull() &\
        ~typ_data.Consumption.isnull()].copy()
    return(typ_data)
```

## 5.1 Weighted mean

$$
\begin{equation}
Wd = 1 \div n \sum_i^n C_i \div D_i \times R_i
\end{equation}
$$


**Equation (7)**

```python
def computeWeightedMean(typ_data, lim=1e7):
    typ_data = truncateData(typ_data, lim)
    C = typ_data["Consumption"]
    D = typ_data["Demand"]
    r = typ_data["Residential"]
    performance = (C / D * r).mean() # <-- Equation (7)
    return(performance)
```

## 5.2 Absolute difference

\begin{equation}
  \Delta = \sum_i \left \| C_i - D_i \right \| 
\end{equation}

**Equation (s1)**

```python
def absoluteDifference(typ_data, lim=1e7):
    typ_data = truncateData(typ_data, lim)
    C = typ_data["Consumption"]
    D = typ_data["Demand"]
    Delta = sum(abs(C - D))    # <-- Equation (s1)
    return(Delta)
```

## 5.3 Percentage of over estimated points

```python
def overEstimated(typ_data, lim=1e7):
    typ_data = truncateData(typ_data, lim)
    C = typ_data["Consumption"]
    D = typ_data["Demand"]
    O = sum(D > C) / len(D)
    return(O)
```

## 5.4 Weighted Z.statistic

$$
\begin{equation}
  Z =1 \div n\sum_i^n \cfrac{d_i-c_i}{\sqrt{ \left \|
  	  \cfrac{c_i-\left(1-c_i\right)}{\sum{D_i \times r_i}}
  	  \right \|
  	  }}
\end{equation}
$$

**Equation (8)**

$$
\begin{equation}
  c_i = \frac{C_i}{\sum_i{C_i}}
\end{equation}
$$


**Equation (9)**

$$
\begin{equation}
  d_i = \frac{D_{i} \times r_{i}}{\sum_i{\left(D_{i} \times r_{i}\right)
  }}
\end{equation}
$$

**Equation (10)**

```python
def computeZ(typ_data, lim=1e7):
    typ_data = truncateData(typ_data, lim)
    C = typ_data["Consumption"]
    D = typ_data["Demand"]
    r = typ_data["Residential"]
    c = C / sum(C)                                               # <-- Equation (09)
    d = D * r / sum(D * r)                                       # <-- Equation (10)
    Z = np.mean((d - c) / np.sqrt(abs((c-(1-c)) / sum(D * r))))  # <-- Equation (08)
    return(Z)
```

## 5.5 Tables

```python
P = {}; Delta = {}; O = {}; Z = {}
for typ_pos in typologies_data:
    typ = typologies_data[typ_pos]
    #typ_name = typ.split("_")[-1].split(".")[0]
    performance = []; difference = []; z = []; o = []
    modes = ["S", "W"]
    for mode in modes:
        typ_data = getData(typ, aggregate="Mean", mode=mode)
        for l in [1e5, 1e6, 1e7]:
            # performance (weighted mean)
            performance_typ = computeWeightedMean(typ_data, lim=l)
            performance.append("{:0.2f}".format(performance_typ))
            # absolute difference
            difference_typ = absoluteDifference(typ_data, lim=l)
            difference.append("{:0.2f}".format(difference_typ / 1e6))
            # over estimated points
            over_typ = overEstimated(typ_data, lim=l)
            o.append("{:0.2f}".format(over_typ))
            # Z-statistic
            z_typ = computeZ(typ_data, lim=l)
            z.append("{:0.2f}".format(z_typ))
            
    P["{} {}".format(typ_pos, typ)] = performance
    Delta["{} {}".format(typ_pos, typ)] = difference
    O["{} {}".format(typ_pos, typ)] = o
    Z["{} {}".format(typ_pos, typ)] = z
```
